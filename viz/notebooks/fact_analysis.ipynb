```python
import matplotlib.pyplot as plt
import numpy as np

# %matplotlib inline

# zeroes = np.zeros(3)
# print(zeroes)

species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Number of penguins by sex')
ax.legend()

plt.show()
```

```python
df.loc[:, ['Name', 'Price/Earnings', 'Industry Avg. (P/E)', 'Implied Downside (P/E)']]
```

```python
def rstrip_x(r: str):
    return int(r.rstrip("x"))
df['P/E'] = [rstrip_x(pe) for pe in df['Price/Earnings']]
df['P/E']
```

```python
df['P/E Industry Avg.'] = [rstrip_x(pe) for pe in df['Industry Avg. (P/E)']]
df['P/E Industry Avg.']
```

```python
def rstrip_percentage(r: str):
    return float(r.rstrip("%"))
df['P/E Implied Downside (in %)'] = [rstrip_percentage(pe) for pe in df['Implied Downside (P/E)']]
df['P/E Implied Downside (in %)']

# df['P/E Implied Downside (in %)'] = [rstrip_percentage(pe) for pe in df['Implied Downside (P/E)']]
```

```python
import matplotlib.pyplot as plt

companies = df['Name']

x = np.arange(len(companies))

ratios = df.loc[:, ['P/E', 'P/E Industry Avg.']]

width = 0.25
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

ax.set_title("Hindenburg's Adani Fact Analysis (Price/Earning)")

for ratio in ['P/E', 'P/E Industry Avg.']:
    offset = width * multiplier
    rects = ax.bar(x + offset, ratios[ratio].to_numpy(), width, label=ratio)
    ax.bar_label(rects, padding=2)
    multiplier += 1

ax.set_ylabel('Price/Earning')
ax.set_xticks(x + .125, companies, rotation=45)
ax.legend(loc='upper right', ncols=2)
ax.set_ylim(0, 1000)

plt.show()


```

```python
import pandas as pd

df = pd.read_csv("../../data/Fundamental_Downside_Data.csv")
# print(df.head())

price_sales_ratio_col = 'Price/Sales'
industry_avg_price_sales_ratio_col = 'Industry Avg. (P/S)'

ev_ebitda_ratio_col = 'EV/EBITDA'
industry_avg_ev_ebitda_ratio_col = 'Industry Avg. (EV/EBITDA)'

cleaned_df = df.copy()

cleaned_df[price_sales_ratio_col] = [float(r.rstrip("x")) for r in cleaned_df[price_sales_ratio_col]]
cleaned_df[industry_avg_price_sales_ratio_col] = [float(r.rstrip("x")) for r in cleaned_df[industry_avg_price_sales_ratio_col]]

cleaned_df[ev_ebitda_ratio_col] = [float(r.rstrip("x")) for r in cleaned_df[ev_ebitda_ratio_col]]
cleaned_df[industry_avg_ev_ebitda_ratio_col] = [float(r.rstrip("x")) for r in cleaned_df[industry_avg_ev_ebitda_ratio_col]]

cleaned_df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# companies = df['Name']

x = np.arange(len(df['Name']))

# df['P/E'] = [rstrip_x(pe) for pe in df['Price/Earnings']]
# df['P/E Industry Avg.'] = [rstrip_x(pe) for pe in df['Industry Avg. (P/E)']]

ratio_cols = [price_sales_ratio_col, industry_avg_price_sales_ratio_col]

ratios = cleaned_df.loc[:, ratio_cols]

width = 0.25
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

ax.set_title("Hindenburg's Adani Fact Analysis (Price/Sales)")

for ratio in ratio_cols:
    # print(ratios[ratio])
    offset = width * multiplier
    rects = ax.bar(x + offset, ratios[ratio], width, label=ratio)
    ax.bar_label(rects, padding=2)
    multiplier += 1

ax.set_ylabel('Price/Sales')
ax.set_xticks(x + .125, df['Name'], rotation=45)
ax.legend(loc='upper right', ncols=2)
ax.set_ylim(0, 150)

plt.show()
```
