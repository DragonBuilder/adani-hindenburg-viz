```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../data/Adani_Holdings_Data.csv")
df=df[:-1]
df.head()
```

```python
one_yr_gain="1-YR Stock % Gain"
df[one_yr_gain]
```

```python
one_yr_gain = "1-Yr Gain (in %)"
three_yr_gain = "3-Yr Gain (in %)"

def rstrip_percentage(num: str) -> int:
    return int(num.rstrip("%"))

df[one_yr_gain] = [rstrip_percentage(gain) for gain in df["1-YR Stock % Gain"]]
df[one_yr_gain]

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.bar(df['Company Name'], df[one_yr_gain])
ax.set_xticks(np.arange(len(df['Company Name'])), df['Company Name'], rotation=45)

ax.set_ylabel('1-Yr Stock % Gain')
ax.set_title('Adani Companies Performance')

plt.show()
```
